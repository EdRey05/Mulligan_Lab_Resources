```python
import ipywidgets as widgets
import ipysheet
```

```python
people = widgets.RadioButtons(options=[('One', 1), ('Two', 2), ('Three', 3)], description='Players:', )
people
```

```python
b = widgets.Text(description='Name')
b
```

```python
c = widgets.IntSlider(min=0, max=10, description='Number:')
c
```

```python
d = widgets.FloatText(description='Score:')
d
```

```python
e = widgets.Checkbox(description='Voted?')
e
```

```python
f = widgets.Combobox(options=['CAD', 'USD', 'MXN', 'EUR'], description='Currency:')
f
```

```python
g = widgets.ToggleButtons(options=['Yes', 'No', 'Maybe'], description='Did you lie?:')
g
```

```python
a = people.value

#Note: not all of these work very well inside the sheet
InteractiveInputs = ipysheet.sheet()

column1 = ipysheet.column(0, [widgets.Text(description='Name') for i in range(a)])
column2 = ipysheet.column(1, [widgets.IntSlider(min=0, max=10, description='Favorite number:') for i in range(a)])
column2 = ipysheet.column(2, [widgets.FloatText(description='Score:') for peps in range(a)])
column3 = ipysheet.column(3, [widgets.Checkbox(description='Voted?') for peps in range(a)])
column4 = ipysheet.column(4, [widgets.Combobox(options=['CAD', 'USD', 'MXN', 'EUR'], description='Currency:') for i in range(a)])
#column5 = ipysheet.column(5, [widgets.ToggleButtons(options=['Yes', 'No'], description='Did you lie?:') for peps in range(a)])

InteractiveInputs
```
